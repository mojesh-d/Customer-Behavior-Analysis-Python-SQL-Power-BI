```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
#create a column age_group
labels=['Young-adult','Ault','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
#create purchase frequency days column
frequency_maping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_maping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']]
```

```python
#check the discount applied column and promo code used column same or not
(df.discount_applied==df.promo_code_used).all()
```

```python
#its trure it means not necesseary to maintain 2 columns so i delate promo_code_used column
df=df.drop('promo_code_used',axis=1)
```

```python
from sqlalchemy import create_engine

# SQL Server connection details
server = r'MOJESH\MOJEE'   # mentioin server name of MS SQL
database = 'Customer_shopping'                  # mention database name where we want to upload the table
driver = 'ODBC Driver 17 for SQL Server'  # driver
# Create connection string
connection_string = f'mssql+pyodbc://@{server}/{database}?driver={driver}&trusted_connection=yes'

# Create SQLAlchemy engine
engine = create_engine(connection_string)

# Upload DataFrame to SQL Server (creates table 'Customer')

df.to_sql('Customer', con=engine, if_exists='replace', index=False)

print("Table uploaded successfully!")
```
